### Test Timestamps and Date and Time transformation
#### Obersvations:
Hackernews uses timestamps
Races has date and time. Letter 'Z' in military time indicates ZULU Time Zone which is equivalent to UTC.

# transform timestamp to time_df
t = pd.to_datetime(df_news['timestamp'], unit='ms')

df_time = pd.DataFrame(columns = ['timestamp', 'hour', 'day', 'week']) 
df_time['timestamp'] = df_news['timestamp']
df_time['hour']= t.dt.hour
df_time['day']= t.dt.day
df_time['week']= t.dt.week
df_time['month']= t.dt.month
df_time['year']=t.dt.year
df_time['weekday']=t.dt.weekday
df_time.head()

```python

# add times from races

from datetime import datetime
import calendar

# Step 1 convert to time string

# 2021-05-23 13:00:00Z
# concat date and time, remove "Z" from time
sample_date = df_races.iloc[0].date
print(sample_date)
sample_time = df_races.iloc[0].time
#remove "Z" label at the end
sample_time_no_z = sample_time[:-1]

print(sample_time)
print(sample_time_no_z)
sample_datetime = sample_date + ' ' + sample_time_no_z
print(sample_datetime)
dt = datetime.strptime(sample_datetime, "%Y-%m-%d %H:%M:%S")
# -> datetime.datetime(2015, 7, 13, 9, 38, 17)

#utc because of "z"
utc_time = calendar.timegm(dt.utctimetuple())
print(utc_time)

#test
# if you encounter a "year is out of range" error the timestamp
# may be in milliseconds, try `ts /= 1000` in that case
print('-------right timestamp?')
print(datetime.utcfromtimestamp(utc_time).strftime('%Y-%m-%d %H:%M:%S'))
```
